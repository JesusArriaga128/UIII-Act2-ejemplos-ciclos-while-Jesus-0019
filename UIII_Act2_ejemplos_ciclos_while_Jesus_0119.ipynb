### Ejemplos de ciclo while

```python
# ejemplo 1
# Descripción: Ciclo que resta gastos a un presupuesto inicial. Usa un if para avisar cuando el dinero se agota.
saldo = 500
while saldo > 0:
    gasto = float(input("Ingresa el monto del gasto: "))
    saldo = saldo - gasto
    if saldo <= 0:
        print("Atención: Te has quedado sin fondos.")
```

```python
# ejemplo 2
# Buscador de Número
# Descripción: Programa que pide datos hasta que se ingresa el valor "100". El if confirma el éxito al usuario.
numero = 0
while numero != 100:
    numero = int(input("Ingresa un número (Adivina el secreto): "))
    if numero == 100:
        print("¡Número correcto encontrado!")
```

```python
# ejemplo 3
# Contador de IA
# Descripción: Ciclo limitado a 5 entradas que usa un if para contar cuántas veces mencionas la palabra "IA".
intentos = 1
cuenta_ia = 0
while intentos <= 5:
    palabra = input("Ingresa una tecnología: ")
    if palabra == "IA":
        cuenta_ia = cuenta_ia + 1
    intentos = intentos + 1
print("Menciones de IA encontradas:", cuenta_ia)
```

```python
# ejemplo 4
# Filtro de Pares
# Descripción: Captura números continuamente y usa un if con el operador % para mostrar solo los que son pares.
valor = 1
while valor != 0:
    valor = int(input("Introduce un número (0 para salir): "))
    if valor % 2 == 0:
        print("El número es par.")
```

```python
# ejemplo 5
# Acumulador con Límite
# Descripción: Suma valores hasta llegar a 1000. El if detecta y advierte si la última entrada superó el límite máximo.
total = 0
while total < 1000:
    monto = float(input("Ingresa cantidad a sumar al total: "))
    total = total + monto
    if total > 1000:
        print("Límite de seguridad superado.")
print("Total final:", total)
```
